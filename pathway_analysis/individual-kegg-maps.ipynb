# Accessing Individual KEGG Maps via GET and `PIL`

Use GET to get individual KEGG maps, with user-selected highlighted nodes. You can highlight certain nodes (metabolites or enzymes) with custom colours, by forming one very long URL string. 

### GET Request

* ref: https://www.kegg.jp/kegg/docs/weblink.html#map2
* Call to `https://www.kegg.jp/kegg-bin/show_pathway`
* Pull `.png` off their servers.
* Accepts KEGG metabolite Ids. Uniprot IDs must be converted to NBI Gene Ids (Entrez). 

### ASCII-encoding
* `%23` is `#`
* `%0d` - carriage return
* `%0a` - line feed (?)

```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import requests

from PIL import Image
import re

def get_kegg_pathway_img(pw_idf , node_colour_dict, verbose=True):
    """GETs a single pathway from KEGG, colouring selected nodes. 
    
    PARAMS
    ------
    pw_id: str; KEGG ID of pathway of interest.
    node_colour_dict: dictionary of kegg IDs of elements of interest as keys (str), and 
    selected colour in hexcode (str). 
    verbose: boolean; verbosity param
    
    RETURNS
    -------
    img: PIL.PngImagePlugin.PngImageFile object of the retrieved pathway, if status code = 200,
    else error.
    
    See also: https://www.kegg.jp/kegg/docs/weblink.html#map2
    """
    url_prefix = "https://www.kegg.jp/kegg-bin/show_pathway?map="+pw+"&multi_query="
    url_suffix = [] # init as list; .join() to a string later
    for k in list(node_colour_dict.keys()):
        node_colour = node_colour_dict[k].replace("#", "%23")
        url_suffix.append(k + "+" + node_colour)
    url_suffix = "%0d%0a".join(url_suffix)
    # create full url
    url = url_prefix + url_suffix
    
    r = requests.get(url)

    try:
        if r.status_code != 200:
            raise ConnectionError
        ms = re.finditer('src="(/tmp/mark_pathway[^"]*.png)"', r.text)
        m = list(ms).pop()
        img = Image.open(requests.get("http://www.kegg.jp%s" % m.group(1), stream=True).raw)
        return img
    except ConnectionError as e:
        print("Error in GET: Status code %s" % r.status_code)
```

```python
# GET pathway using some made-up data
pw = "hsa00010"
# Colours are either red (#ff0000) or green (#a6d854)
elements_dict = {"hsa00010": {"7167": "#ff0000", "C00118":"#a6d854", "C00036":"#ff0000"}, 
                 "hsa00020": {"C00158": "#ff0000", "5105": "#ff0000", "C00036": "#a6d854"}
                }

img = get_kegg_pathway_img(pw, elements_dict[pw])
```
