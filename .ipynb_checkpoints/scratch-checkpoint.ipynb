```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=3)

import numpy as np
import pandas as pd
```

```python
d0 = pd.read_csv("~/Documents/my_vignettes/data/lcms-norm-eval-results.csv")
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, figsize=(40, 20), sharey=True)

d_t = d0.loc[d0["Metabolites_ID"]=="MTBLS28(ESI+)"]
colnames = ["method"] + [str(90 * x) for x in np.arange(1, 11)]
m0 = d_t[[str(90 * x) for x in np.arange(1, 11)]].values
hm0 = sns.heatmap(m0, annot=True, 
                  annot_kws={"size": 35}, 
                  yticklabels=d_t["method"].values, 
                  xticklabels=[str(90 * x) for x in np.arange(1, 11)], 
                  cbar=False, 
                  ax = ax0
                 )
ax0.set_title("MTBLS28(ESI+)")

d_t = d0.loc[d0["Metabolites_ID"]=="MTBLS28(ESI_)"]
m1 = d_t[[str(90 * x) for x in np.arange(1, 11)]].values
hm1 = sns.heatmap(m1, annot=True, 
                  annot_kws={"size": 35}, 
                  yticklabels=d_t["method"].values, 
                  xticklabels=[str(90 * x) for x in np.arange(1, 11)], 
                  cbar=False, 
                  ax = ax1
                 )
ax1.set_title("MTBLS28(ESI-)")

plt.subplots_adjust(wspace=0.01, hspace=0)

fig.savefig('lc-norm-eval.png', bbox_inches = 'tight', pad_inches = 0)
```
